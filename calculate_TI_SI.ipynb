```python
import utils

cap = utils.openVideoStream('https://www.youtube.com/watch?v=3odlJeCwMxk')
```

```python
utils.extractFrames720p(cap, 30, 5, 30, './frames')
```

```python
ti = utils.calculateTI('./frames')
si = utils.calculateSI('./frames')
optic_flow = utils.calculateOpticalFlow('./frames')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(si, label="SI", marker='o')
plt.xlabel("Frame Index")
plt.ylabel("Spaital Information (SI)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(ti, label="TI (absdiff std)", marker='o')
plt.plot(optic_flow, label="Optical Flow (Lucas-Kanade)", marker='x')
plt.xlabel("Frame Index")
plt.ylabel("Motion Intensity")
plt.title("TI vs Optical Flow (Lucas-Kanade)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
