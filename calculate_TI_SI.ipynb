# Temporal Information, Spatial Information 계산

- TI: 영상이 얼마나 빠른지를 측정
- SI: 영상이 얼마나 복잡한지를 측정

```python
import utils

ti = utils.calculateTI('./clips/clip_004')
si = utils.calculateSI('./clips/clip_004')
optic_flow = utils.calculateOpticalFlow('./clips/clip_004')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(si, label="SI", marker='o')
plt.xlabel("Frame Index")
plt.ylabel("Spaital Information (SI)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"SI Mean: {sum(si) / len(si)}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(ti, label="TI (absdiff std)", marker='o')
plt.plot(optic_flow, label="Optical Flow (Lucas-Kanade)", marker='x')
plt.xlabel("Frame Index")
plt.ylabel("Motion Intensity")
plt.title("TI vs Optical Flow (Lucas-Kanade)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"TI Mean: {sum(ti) / len(ti)}")
print(f"OF Mean: {sum(optic_flow) / len(optic_flow)}")
```
